Here is a raw code example (in R) that is a good start for the normal distribution, but would not get full

Import libraries that will be used.

```python
import numpy as np
import scipy.stats as stats
import pylab as pl
```

Define variables

```python
N = 1000
N1 = 50
x = np.random.normal(size=N)
x1 = np.random.normal(size=N1)
```

Define variables for the x and y axis of the histogram

```python
x_fit = np.arange(min(x),max(x), 1/N)
x1_fit = np.arange(min(x1),max(x1), 1/N1)

y_fit = stats.norm.pdf(x_fit,np.mean(x),np.std(x))
y1_fit = stats.norm.pdf(x1_fit,np.mean(x1),np.std(x1))
```

Plot the histogram and fitted normal curve

```python
pl.subplot(1,2,1)
pl.plot(x_fit, y_fit)
pl.title('Histogram of x')
pl.ylabel('Density')
pl.hist(x,density=True)

#second modified graph x1
pl.subplot(1,2,2)
pl.plot(x1_fit, y1_fit)
pl.title('Histogram of x1')
pl.ylabel('Density')
pl.hist(x1,density=True, bins='rice',histtype='barstacked', color = 'pink')


pl.tight_layout()
pl.show()
```
